```python
def calculate_mean():
    num_st = int(input("enter number of student"))
    grades=[]
    for i in range(num_st):
        grade=float(input("enter the grade"))
        grades.append(grade)
    if num_st==0:
        print("no students")
    else:
        print(f"the average of grade ={sum(grades)/num_st}")  

calculate_mean()             
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mult(a,b):
    return a*b
def div(a,b):
    if b==0:
        print("invalid division")
    else:
        return a/b

while True:
    print("1- add / 2-sub / 3-mult / 4-div /0- Exit ")
    choice=input("enter your choice")
    a=float(input("Enter your num"))
    b=float(input("Enter your num"))
    if choice==1:
        print(add(a,b))
    elif choice ==2:
         print(sub(a,b))           
    elif choice ==3:
         print(mult(a,b))
    elif choice ==4:
         print(div(a,b))
    elif choice ==0:
         break
    else:
        print("invaild operation")    
```
