## task 1 Email Processing

```python
def process_email(email):
    if email.count('@') != 1:
        return "Invalid email"
    
    parts = email.split('@')
    username = parts[0]
    domain_part = parts[1]
    
    if '.' not in domain_part:
        return "Invalid email"
    print(f"Username: {username}")
    last_dot_index = domain_part.rfind('.')
    domain_name = domain_part[:last_dot_index]
    print(f"Domain Name: {domain_name}")
    extension = domain_part[last_dot_index:]
    
    if extension == ".com":
        return "Commercial Domain"
    elif extension == ".edu":
        return "Educational Domain"
    else:
        return "Other Domain"
    
email_input = "Amit_ml@gmail.edu" 
result = process_email(email_input)
print(f"Result: {result}")
```

## task2 Decoding Message

```python
text = "###!!@mocleW EPGTQ!!!6789"
core = text[6:19] 
parts = core.split() 
word1 = parts[0]
word2 = parts[1]

decoded_word1 = word1[::-1] 
new_word = word2.replace("E", "")

print(decoded_word1 + " " + new_word)
```

## task3 Decoding Message

```python
text = "&&&**$gnirts PLIO!!@1234"
core = text[6:17] 
parts = core.split()
word1 = parts[0]
word2 = parts[1]

decoded_word1 = word1[::-1] 
word2 = word2.replace("I", "E")
word2 = word2.replace("O", "U")

print(decoded_word1.capitalize() + " " + word2)
```
