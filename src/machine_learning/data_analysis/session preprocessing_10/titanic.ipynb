```python
for col in num_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR= Q3-Q1
    Lower_Fence= Q1-1.5*IQR
    Upper_Fence= Q3+1.5*IQR
    print(Q1)
    print(Q3)
    print(IQR)
    print(Lower_Fence)
    print(Upper_Fence)
```

```python
Lower_Outliersa=df[df[col] <Lower_Fence][col].values
Lower_Outliersa
```

```python
Upper_Outliersa=df[df[col]> Upper_Fence][col].values
Upper_Outliersa
```

```python
#df = df[(df[col] >= Lower_Fence) & (df[col] <= Upper_Fence)]
```

```python
df[col].replace(Lower_Outliersa,Lower_Fence,inplace=True)
df[col].replace(Upper_Outliersa,Upper_Fence,inplace=True)
```
