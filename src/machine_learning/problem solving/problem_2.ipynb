```python
class BanjoChecker:
    def __init__(self, name):
        self.name = name  
    def check_if_plays(self):
        if self.name[0].lower() == 'r':
            return f"{self.name} plays banjo"
        else:
            return f"{self.name} does not play banjo"

# --- Execution ---
player1 = BanjoChecker("Ringo")
player2 = BanjoChecker("Martin")

print("OOP Output 1:", player1.check_if_plays()) # plays banjo
print("OOP Output 2:", player2.check_if_plays()) # does not play banjo
```
