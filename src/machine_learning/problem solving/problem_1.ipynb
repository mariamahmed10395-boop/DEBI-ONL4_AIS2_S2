```python
class WordReverser:
      def __init__(self, sentence):
        self.sentence = sentence  

      def reverse_words(self):
        words = self.sentence.split()
        reversed_words = words[::-1]

        return " ".join(reversed_words)

# --- Execution ---
reverser = WordReverser("The greatest victory is that which requires no battle")
print("OOP Output:", reverser.reverse_words())
# 
```
